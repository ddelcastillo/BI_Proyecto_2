# Procesamiento de datos para Análisis 3: Flujo de aéreo y terrestre

Se realizará una nueva tabla de hechos que condense los 3 tipos de infraestructura, paeaje, aeropuerto origen y aeropuerto destino. 

## Datos aeropuertos

```python
import pandas as pd
fact_aero = pd.read_csv("datos modificados/aeropuertos_historico_v3.csv", index_col=0)
```

```python
fact_aero.head()
```

```python
aero = pd.read_csv("datos modificados/aeropuertos_v3.csv", index_col=0)
```

```python
aero.head()
```

## Eliminación de datos de trafico aéreo anteriores a 2014

```python
fact_aero = fact_aero[fact_aero.cod_fecha >= "2014_01"]
print('Cantidad de datos con desdes 2014:',len(fact_aero.index))
```

```python
min = fact_aero['cod_fecha'].min()
max = fact_aero['cod_fecha'].max()

print('Min', min, 'Max', max)
fact_aero.head()
```

## Creación de nueva dimension Infraestructura

Es importante mencionar que la columna tipo tiene 3 valores posibles. "aero_origen", "aero_destino", "peaje"

```python
infra1 = aero.filter(['sigla','nombre','departamento'], axis=1)
infra1['sigla'] = infra1['sigla'] +'_O'
infra2 = aero.filter(['sigla','nombre','departamento'], axis=1)
infra2['sigla'] = infra2['sigla'] +'_D'
infra1.head()
```

```python
infra1['tipo'] = 'aero_origen'
infra1.head()
```

```python
infra2['tipo'] = 'aero_destino'
infra2.head()
```

```python
frames = [infra1, infra2]

infra = pd.concat(frames)
infra.head()
```

```python
infra.tail()
```

```python
print(infra['sigla'].value_counts())

print(infra['sigla'].isna().sum())
```

```python
infra = infra.rename(columns={'sigla': 'codigo'})
infra.head()
```

## Creacción de tabla de hechos

```python
fact_infra1 = fact_aero.groupby(['cod_origen', 'cod_destino','cod_fecha']).agg({'pasajeros_transportados': 'sum'}).reset_index()
fact_infra1.tail()
print(fact_infra1[['cod_origen', 'cod_destino','cod_fecha']].value_counts())
```

```python
min = fact_aero['pasajeros_transportados'].min()
max = fact_aero['pasajeros_transportados'].max()

print('Min', min, 'Max', max)
```

```python
fact_infra2 = fact_infra1.filter(['cod_origen','cod_fecha','pasajeros_transportados', 'cod_destino'], axis=1)
fact_infra2['ruta'] = fact_infra2['cod_origen'] + fact_infra2['cod_destino']
fact_infra2['cod_origen'] = fact_infra2['cod_origen'] +'_O'
fact_infra2 = fact_infra2.rename(columns={'cod_origen': 'codigo_infra'})
fact_infra2.head()
```

```python
fact_infra3 = fact_infra1.filter(['cod_destino','cod_fecha','pasajeros_transportados', 'cod_origen'], axis=1)
fact_infra3['ruta'] = fact_infra3['cod_origen'] + fact_infra3['cod_destino']
fact_infra3['cod_destino'] = fact_infra3['cod_destino'] +'_D'
fact_infra3 = fact_infra3.rename(columns={'cod_destino': 'codigo_infra'})

fact_infra3.head()
```

```python
frames = [fact_infra2, fact_infra3]
fact_infra = pd.concat(frames)
fact_infra = fact_infra.rename(columns={'pasajeros_transportados': 'cantidad_personas'})

fact_infra = fact_infra.filter(['codigo_infra','cod_fecha', 'ruta', 'cantidad_personas'], axis=1)

fact_infra.tail()
```

```python
print(fact_infra['codigo_infra'].isna().sum())
fact_infra.head()
```

## Agregar peajes a la dimensión infraestructura


```python
peajes = peajes.filter(['nombre','departamento'], axis=1)
peajes['codigo'] = peajes['nombre'] 
peajes['tipo'] = 'peaje'
peajes['ruta'] = '_P'
cols = list(peajes.columns.values)
print(cols)
```

```python
frames = [infra, peajes]

infra = pd.concat(frames)
infra.head()
```

## Aproximación de personas por carro
A partir del siguiente link, se estima que el promedio de personas por carro es de 2.2, tomando (40+60*3 )/100
https://www.eltiempo.com/bogota/seis-de-cada-diez-personas-dijeron-compartir-su-vehiculo-en-bogota-90186

```python
fact_peajes['cantidad_personas'] = 2.2*fact_peajes['cantidad_personas']
fact_peajes.head()
```
