### Remover inconsistencias en la tabla de históricos

```python
aero_sigle = aero['sigla']
origen_column = historico['origen']
destino_column = historico['destino']

# peajes = peajes.loc[peajes['nombre'].isin(iterable_recaudo)]

origen_vector = historico['origen'].isin(aero_sigle)
destino_vector = historico['destino'].isin(aero_sigle)

ambas = origen_vector & destino_vector
historico_coherente = historico.loc[ambas]

print('El número de registros en históricos con '
      'datos de origen y destino en aeropuertos es: '
      '',ambas.values.sum())
print('El número de registros a eliminar de históricos por'
      'inconsistencias es: ',historico.shape[0] - ambas.values.sum()) #Registros a eliminar en histórico

print(len(origen_vector), len(destino_vector))
print(historico_coherente.shape[0])
```

### Códigos de aeropuertos en históricos sin entradas en la tabla de aeropuertos


```python
origen_incoherente = historico.loc[~origen_vector]['origen']
destino_incoherente = historico.loc[~destino_vector]['destino']

columna_incoherente = pd.concat([origen_incoherente,
                                 destino_incoherente],
                                axis=0, ignore_index=True)

print('Número de aeropuertos en tabla de aeropuertos'
      '', aero.shape[0])
print('Número de códigos únicos de aeropuertos '
      'en histórico sin referencia a tabla aeropuertos: '
      '', len(columna_incoherente.unique()))

codigos_totales = pd.concat([historico['origen'],
                                 historico['destino']],
                                axis=0, ignore_index=True)
print('Número de códigos únicos de aeropuertos '
      'en histórico: '
      '', len(codigos_totales.unique()))
```

## Crear tabla fecha


```python
historico_v2 = pd.read_csv('./datos_modificados/aeropuertos_historico_v2.csv')

min = historico_v2['cod_fecha'].min()
max = historico_v2['cod_fecha'].max()

print('Min', min, 'Max', max)
print(min[5:])

print(2//4 + 1)
```

```python
x = []
meses = ['Enero','Febreo','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']
m = 0;

anioInit, mesInit = int(min[0:4]), int(min[5:])
anioFin,  mesFin = int(max[0:4]), int(max[5:])

# print((anioFin-anioInit)*12+mesFin+(12-mesInit+1))

for i in range(1, (anioFin-anioInit-1)*12+mesFin+(12-mesInit+1)+1):
    anio = anioInit + (i-mesInit+1)//12
    mesNumber = (i-1)%12 + 1
    mes = meses[mesNumber - 1]
    semestre = 1 if mesNumber <=6 else 2
    cuad = 1 if mesNumber <= 4 else (2 if mesNumber <=8 else 3)
    tri = (mesNumber-1)//3 +1
    cod_fecha = f'{anio}_{mesNumber}'
    x.append([cod_fecha,anio,semestre,cuad,tri,mes, mesNumber])

fechas = pd.DataFrame(x, columns = ['Cod_fecha', 'Anio', 'Semestre', 'Cuatrimestre','Trimestre', 'Mes', 'Mes_numerico'])
fechas.head(15)

fechas.to_csv('./datos_modificados/fechas_v1.csv')
```
